```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score, classification_report
import numpy as np
```

```python
print(df.head())

# Display basic information about the dataset
print(df.info())

# Check for missing values
print(df.isnull().sum())

# Drop columns with all NaN values
df.dropna(axis=1, how='all', inplace=True)

# Summary statistics
print(df.describe())

# Visualize the distribution of the target attributes
sns.countplot(x='Rank-1\n(Target Attribute-1)\nFor Binary Classification', data=df)
plt.title('Distribution of Rank-1')
plt.show()

sns.countplot(x='Rank-2\n(Target Attribute-2)\nFor Binary Classification', data=df)
plt.title('Distribution of Rank-2')
plt.show()

# Correct column names based on the actual column names in the dataset
target_1 = 'Rank-1\n(Target Attribute-1)\nFor Binary Classification'
target_2 = 'Rank-2\n(Target Attribute-2)\nFor Binary Classification'
multi_class_1 = 'Total practice score\n(Target Attribute-1)\nFor Multi-Class Classification'
multi_class_2 = 'Total knowledge score\n(Target Attribute-2)\nFor Multi-Class Classification'

# Encode categorical variables
le = LabelEncoder()
categorical_columns = ['Group', 'Patient occupation', 'sex', 'lived/worked with dengue patient', 'geographical location of the house', 'Residence type']
for col in categorical_columns:
    df[col] = le.fit_transform(df[col])

# Check if the column name 'Age category' exists in the DataFrame
if 'Age category' not in df.columns:
    raise ValueError("Column 'Age category' not found in the DataFrame")

# Encode age categories
age_mapping = {
    '16-30 years old': 1,
    '31-45 years old': 2,
    '46-60 years old': 3,
    '60+ years old': 4
}
df['Age category'] = df['Age category'].map(age_mapping)

# Scaling numerical variables
scaler = StandardScaler()
numerical_columns = ['Age category']
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Handle missing values
imputer = SimpleImputer(strategy='mean')
df = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

# Split the data
train, temp = train_test_split(df, test_size=0.3, random_state=42)
val, test = train_test_split(temp, test_size=1/3, random_state=42)

print(f'Training set size: {train.shape}')
print(f'Validation set size: {val.shape}')
print(f'Test set size: {test.shape}')
```

```python
# Define function to train and evaluate SVM model
def train_and_evaluate_svm(X_train, y_train, X_val, y_val, X_test, y_test, target_name):
    # Initialize SVM model
    svm_model = SVC(kernel='linear', random_state=42)  # You can choose different kernels like 'rbf', 'poly', etc.

    # Fit the model on the training data
    svm_model.fit(X_train, y_train)

    # Predictions on the validation set
    val_predictions = svm_model.predict(X_val)

    # Evaluate the model
    print(f"Validation Accuracy ({target_name}):", accuracy_score(y_val, val_predictions))
    print(f"Validation Classification Report ({target_name}):")
    print(classification_report(y_val, val_predictions))

    # Predictions on the test set
    test_predictions = svm_model.predict(X_test)

    # Evaluate the model on test set
    print(f"Test Accuracy ({target_name}):", accuracy_score(y_test, test_predictions))
    print(f"Test Classification Report ({target_name}):")
    print(classification_report(y_test, test_predictions))

# For Target Attribute 1 (Rank-1)
X_train_1 = train.drop(columns=[target_1, target_2])  # Drop both target columns
y_train_1 = train[target_1]
X_val_1 = val.drop(columns=[target_1, target_2])
y_val_1 = val[target_1]
X_test_1 = test.drop(columns=[target_1, target_2])
y_test_1 = test[target_1]

print("SVM Model for Rank-1 (Target Attribute-1):")
train_and_evaluate_svm(X_train_1, y_train_1, X_val_1, y_val_1, X_test_1, y_test_1, target_1)

# For Target Attribute 2 (Rank-2)
X_train_2 = train.drop(columns=[target_1, target_2])  # Drop both target columns
y_train_2 = train[target_2]
X_val_2 = val.drop(columns=[target_1, target_2])
y_val_2 = val[target_2]
X_test_2 = test.drop(columns=[target_1, target_2])
y_test_2 = test[target_2]

print("\nSVM Model for Rank-2 (Target Attribute-2):")
train_and_evaluate_svm(X_train_2, y_train_2, X_val_2, y_val_2, X_test_2, y_test_2, target_2)
```

```python
# Define function to plot ROC curve and calculate AUC using decision function or predict_proba
from sklearn.metrics import roc_curve, auc
def plot_roc_curve(model, X_test, y_test, target_name):
    # Check if the model supports decision_function
    if hasattr(model, "decision_function"):
        # Calculate decision function scores
        y_score = model.decision_function(X_test)
    elif hasattr(model, "predict_proba"):
        # Calculate probability estimates
        y_score = model.predict_proba(X_test)[:, 1]
    else:
        raise AttributeError("Model does not have decision_function or predict_proba method.")

    # Calculate fpr, tpr, and thresholds
    fpr, tpr, thresholds = roc_curve(y_test, y_score)

    # Calculate AUC
    roc_auc = auc(fpr, tpr)

    # Plot ROC curve
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (AUC = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--', label='Random Guessing')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate (FPR)')
    plt.ylabel('True Positive Rate (TPR)')
    plt.title(f'ROC Curve for {target_name}')
    plt.legend(loc='lower right')
    plt.grid(True)
    plt.show()

# Plot ROC curve and calculate AUC for SVM models
plot_roc_curve(svm_model_1, X_test_1, y_test_1, target_1)
plot_roc_curve(svm_model_2, X_test_2, y_test_2, target_2)

# Plot ROC curve and calculate AUC for KNN models
plot_roc_curve(knn_model_1, X_test_1, y_test_1, target_1)
plot_roc_curve(knn_model_2, X_test_2, y_test_2, target_2)

# Plot ROC curve and calculate AUC for Random Forest models
plot_roc_curve(rf_model_1, X_test_1, y_test_1, target_1)
plot_roc_curve(rf_model_2, X_test_2, y_test_2, target_2)

# Plot ROC curve and calculate AUC for Naive Bayes models
plot_roc_curve(nb_model_1, X_test_1, y_test_1, target_1)
plot_roc_curve(nb_model_2, X_test_2, y_test_2, target_2)

# Plot ROC curve and calculate AUC for Decision Tree models
plot_roc_curve(dt_model_1, X_test_1, y_test_1, target_1)
plot_roc_curve(dt_model_2, X_test_2, y_test_2, target_2)

# Plot ROC curve and calculate AUC for Adaboost models
plot_roc_curve(adaboost_model_1, X_test_1, y_test_1, target_1)
plot_roc_curve(adaboost_model_2, X_test_2, y_test_2, target_2)

# Plot ROC curve and calculate AUC for XGBoost models
plot_roc_curve(xgb_model_1, X_test_1, y_test_1, target_1)
plot_roc_curve(xgb_model_2, X_test_2, y_test_2, target_2)
```
